```python
#!pip install scikit-optimize
from skopt import BayesSearchCV
from skopt.space import Real,Categorical,Integer

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("/content/merged-csv.csv")  #you can try it on any dataset but only on categorical dataset
df
```

```python
df = df.select_dtypes(exclude=['object']) #drop all string columns
```

```python
X=df.drop(columns=["is_fraud"])
Y=df["is_fraud"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

#SVM

```python
sc={
    'C':Real(1e-2,1e+2,prior="log-uniform"),  #u can also write like this   'C': Integer(1, 10),
    'gamma':Real(1e-2,1e+2,prior="log-uniform"), #u can also write like this   'gamma': Integer(1, 10),
    'kernel':Categorical(['linear','rbf'])
}
```

```python
svc=SVC()
```

```python
bs=BayesSearchCV(estimator=svc,search_spaces=sc,n_iter=10,cv=5,random_state=0)
```

```python
bs.fit(X_train,Y_train)
print(bs.best_params_)
```
